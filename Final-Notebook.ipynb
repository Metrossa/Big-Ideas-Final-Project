As for the first dataset, we are using it to figure out which neighborhood has the most fire safety so you don't have to worry about your car catching on **fire** when parking it.

## The Best Neighborhood

In order to determine which neighborhood is the best for parking your car, we are first going to analyze each of the datasets and draw conclusions from them.

### 1. Fire Incidents Dataset

In the code below, we analyzed which neighborhood has the least amount of fire incidents by counting how many times each neighborhood appeared in the dataset and then displaying the data in the graph above. As you can see, Mount Oliver Borough has had only 7 fire incidents since 2013 (which is the year the dataset begins at), making it the safest neighborhood to park your car in according to this submetric

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load the dataset
df = pd.read_csv("fires.csv")

# group the data by the number of times the neighborhood is mentioned
neighborhood_counts = df.groupby('neighborhood').size()

# sort the data and show the last ten
neighborhood_counts = neighborhood_counts.sort_values(ascending=False).tail(10)

# create a bar graph
plt.barh(neighborhood_counts.index, neighborhood_counts.values)

plt.title("Top 10 Safest Neighborhoods to Park Your Car in Pittsburgh")
plt.xlabel("Number of Fire Incidents")
plt.ylabel("Neighborhood")

plt.show()
```

### 2. Aggregated Parking Dataset

In the code below, we analyzed which neighborhood has the least amount of parking meter transactions because that means that there are not that many parking meters around, meaning that the likelihood of you getting a parking ticket when parking in that neighborhood. As you can see, Squirell Hill has the least amount of meter transactions making it the best neighborhood to park your car in according to this submetric.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('parking1.csv')

# grouping the data by zone and then sum the meter_transactions column
zone_totals = df.groupby('zone')['meter_transactions'].sum() 

# sort the data and only select the top 10 zones
top_zones = zone_totals.sort_values(ascending=False).head(10)

# create a bar chart 
plt.bar(top_zones.index, top_zones.values)
plt.xlabel('Zone')
plt.ylabel('Total Meter Transactions')
plt.title('Lowest 10 Zones by Meter Transactions')
plt.xticks(rotation=90)

plt.show()
```

### 3. Condemned and Dead-End Properties

In the code below, we analyzed which neighborhood has the least amount of condemned and dead-end properties to ensure that your car will be safe whenever it is parked in this neighborhood. As you can see, Carnegie has the least amount of condemned and dead-end properties making it the safest neighborhood to park your car in according to this submetric.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# laod the dataset
df = pd.read_csv("dead.csv")

# get the zipcode from the address column
df['zipcode'] = df['address'].str.extract('(\d{5})$')

# group by the number of times the zipcode is mentioned
hoods = df.groupby(df['zipcode']).size()

#sort the data and show the bottom 10
hoods = hoods.sort_values(ascending=False).tail(10)

# create a bar graph
plt.barh(hoods.index, hoods.values)

plt.title("Top 10 Safest Zipcodes to Park Your Car in Pittsburgh")
plt.xlabel("Number of Condemned or Dead End Properties")
plt.ylabel("Zipcode")

plt.show()
```
