# Aggregated Parking Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# load the original dataset
df = pd.read_csv('parking1.csv')

# group the data by zone and sum the meter_transactions column
zone_totals = df.groupby('zone')['meter_transactions'].sum() 

# sort the data in descending order and select the top 10 zones
top_zones = zone_totals.sort_values(ascending=False).head(10)

# create a bar chart to display the data
plt.bar(top_zones.index, top_zones.values)
plt.xlabel('Zone')
plt.ylabel('Total Meter Transactions')
plt.title('Lowest 10 Zones by Meter Transactions')
plt.xticks(rotation=90)

# show the plot
plt.show()

```

```python
bottom_zones = zone_totals.sort_values(ascending=False).tail(10)

# create a bar chart to display the data
plt.bar(bottom_zones.index, bottom_zones.values)
plt.xlabel('Zone')
plt.ylabel('Total Meter Transactions')
plt.title('Highest 10 Zones by Meter Transactions')
plt.xticks(rotation=90)

# show the plot
plt.show()
```
